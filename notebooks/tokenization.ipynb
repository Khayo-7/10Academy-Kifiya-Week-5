```python
import os
import sys
sys.path.append('..')

from scripts.data_utils.loaders import *
from scripts.modeling.tokenizer import *
from scripts.modeling.model_utils import *

%autoreload 2
```

```python
DATA_PATH = os.path.join('..', 'resources', 'data')
labeled_dir = os.path.join(DATA_PATH, 'labeled')

tokenized_aligned_dir = os.path.join(DATA_PATH, 'tokenized_aligned')
os.makedirs(tokenized_aligned_dir, exist_ok=True)

tokenized_aligned_filename = 'tokenized_aligned_messages'
filename = os.path.join(labeled_dir, 'labeled_messages.conll')
```

```python
# Load CoNLL file
labeled_data = load_data(filename)
labeled_data.info
```

```python
labeled_data.describe()
```

```python
tokens, labels = separate_tokens_and_labels(labeled_data)
label_id_mapping = {'O': 0, 'B-Product':1, 'I-Product': 2, 'B-LOC': 3, 'I-LOC': 4, 'B-PRICE': 5, 'I-PRICE': 6}
len(tokens), len(labels)
```

```python
model_checkpoint = "bert-base-uncased"
model_checkpoint = "distilbert-base-uncased"
model_checkpoint = "xlm-roberta-base"

auto_tokenizer = initialize_tokenizer(model_checkpoint)
tokenizer = Tokenizer(auto_tokenizer, label_id_mapping=label_id_mapping)

auto_aligned_tokens, auto_aligned_labels, _ = tokenizer.tokenize_and_align_labels(tokens, labels)
auto_aligned_tokens, auto_aligned_labels
```

```python
# Display the first 20 results
for token, label in zip(auto_aligned_tokens[:20], auto_aligned_labels[:20]):
    print(f"{token:20} {label}")
```

```python
# Saving the final tokens and labels to a CoNLL and CSV/JSON file
save_labels(auto_aligned_tokens, auto_aligned_labels, f'{tokenized_aligned_filename}_auto', tokenized_aligned_dir, save_json=True)
```

```python
auto_aligned_tokens_hf, auto_aligned_labels_hf, tokenized_tokens_hf = tokenizer.tokenize_and_align_labels(tokens, labels, use_hf=True)
auto_aligned_tokens_hf, auto_aligned_labels_hf, tokenized_tokens_hf
```

```python
# Display the first 20 results
for token, label in zip(auto_aligned_tokens_hf[:20], auto_aligned_labels_hf[:20]):
    print(f"{token:20} {label}")
```

```python
# Saving the final tokens and labels to a CoNLL and CSV/JSON file
save_labels(auto_aligned_tokens_hf, auto_aligned_tokens_hf, f'{tokenized_aligned_filename}_auto_hf', tokenized_aligned_dir, save_json=True)
```

```python
# Initialize the Amharic segmenter
segmenter = initialize_amharic_segmenter()
am_tokenizer = Tokenizer(segmenter)

am_aligned_tokens, am_aligned_labels, _ = am_tokenizer.tokenize_and_align_labels(tokens, labels)
am_aligned_tokens, am_aligned_labels
```

```python
# Output the first 20 aligned tokens and labels
for token, label in zip(am_aligned_tokens[:20], am_aligned_labels[:20]):
    print(f"{token}: {label}")
```

```python
# Saving the final tokens and labels to a CoNLL and CSV/JSON file
save_labels(am_aligned_tokens, am_aligned_labels, f'{tokenized_aligned_filename}_amseg', tokenized_aligned_dir, save_json=True)
```
